```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window
spark = SparkSession.builder.appName("lrng1").getOrCreate()
```

```python
data=[('Alice',30,'Hyderabad'),('Bob',25,'Pune'),('Carol',29,'Hyderabad'),('Dave',35,'Pune')]
columns=['Name','Age','City']
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df_avg_age=df.groupBy("City").agg(avg("Age").alias("AvgAge"))
df_avg_age.show()
```

```python
df.select("City","Age").groupBy("City").agg(avg("Age").alias("Avg_Age")).show()
```
